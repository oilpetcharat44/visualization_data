Visualization Data

```
import seaborn as sns

sns.set(style="white", color_codes=True)
sns.FacetGrid(df, hue="tagable", size=5) \
   .map(plt.scatter, "word_count", "edu_count") \
   .add_legend()
```

```
sns.FacetGrid(df, hue="tagable", size=6) \
   .map(sns.kdeplot, "word_count") \
   .add_legend()
```

```
sns.FacetGrid(df, hue="tagable", size=6) \
   .map(sns.kdeplot, "edu_count") \
   .add_legend()
```
